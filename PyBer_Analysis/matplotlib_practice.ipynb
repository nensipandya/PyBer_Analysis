```python
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "March", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in us dollars accumulates for each month.
y_axis = [10.02, 23.24, 45.89, 39.20, 35.42, 27.04, 43.87, 10.56, 11.85, 20.71, 20.00, 12.78]
```

```python
# Create the plot
#plt.plot(x_axis, y_axis)

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="r", linewidth=2, label='Boston')

# Create label for x and y axis.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# set the y limit to 0 to 45
plt.ylim(0,45)
# Create a title
plt.title("PyBer Fare By Month")

# Add a grid.
plt.grid()

# Add a alegend.
plt.legend()
```

```python
# Create bar chaet
plt.bar(x_axis, y_axis)
plt.show()
```

```python
# Create a plt
plt.bar(x_axis, y_axis, color="red", label="Boston")

# Create a label for x-axis and y-axis.
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.title("PyBer fare By Month")

# Add Legend.
plt.legend()
```

```python
# Create horizontal bar chart
plt.barh(x_axis, y_axis)
```

```python
# set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "March", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in us dollars accumulates for each month.
y_axis = [10.02, 23.24, 45.89, 39.20, 35.42, 27.04, 43.87, 10.56, 11.85, 20.71, 20.00, 12.78]
plt.plot(x_axis, y_axis ,'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
plt.scatter(x_axis, y_axis, s = [i *3 for i in y_axis])
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, marker='o', color='blue', s = [i*5 for i in y_axis], edgecolor='black', linewidth=2, label='Boston')
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare By Month")
ax.set_ylim(0, 50)
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, marker='o', color='blue', s = [i*5 for i in y_axis], edgecolor='black', linewidth=3, label='Bosston')
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```
